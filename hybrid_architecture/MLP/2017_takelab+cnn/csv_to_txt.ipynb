```python
import pandas as pd
```

```python
vals=[]
txt= open('sts_train.txt', 'w')
txt1= open('sts_train_scores.txt','w')
with open('sts-train.csv','r') as file:
    lines = file.readlines()
    for line in lines:
        try:
            vals.append(line.strip().split('\t'))
            t= line.strip().split('\t')
            txt.write(t[5]+'\t'+t[6]+'\n')
            txt1.write(t[4]+'\n')
        except :
            print(line)
txt.close()
txt1.close()
```
